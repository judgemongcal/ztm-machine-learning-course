```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x,y); # adds some data
type(fig), type(ax)
```

## Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(20,10)) # figsize dictates the dims of the figure

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("sample-plot.png")
```
