## Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(20,10)) # figsize dictates the dims of the figure (width, height)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("sample-plot.png")
```

## Making figures with NumPy Arrays

* Line Plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplots

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use the same data to make a scatter plot
fig,ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter = {"Almond Butter" : 10,
              "Peanut Butter": 8,
              "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter.keys(), nut_butter.values());
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter.keys()), list(nut_butter.values())); # must be in list type
```

```python
# Make some data for Histograms and Plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for Subplots

```python
# Subplot Option 1
fig, ((ax0, ax1),(ax2, ax3)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))

# Plot to each different axis
ax0.plot(x, x /2);
ax1.scatter(np.random.random(10), np.random.random(10));
ax2.bar(nut_butter.keys(), nut_butter.values());
ax3.hist(np.random.randn(100));
```

```python
# Subplot Option 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize=(10,5))

# Plot to each different axis
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter.keys(), nut_butter.values());
ax[1,1].hist(np.random.randn(100));
```
