## Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(20,10)) # figsize dictates the dims of the figure (width, height)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("sample-plot.png")
```

## Two options for Subplots

```python
# Subplot Option 1
fig, ((ax0, ax1),(ax2, ax3)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))

# Plot to each different axis
ax0.plot(x, x /2);
ax1.scatter(np.random.random(10), np.random.random(10));
ax2.bar(nut_butter.keys(), nut_butter.values());
ax3.hist(np.random.randn(100));
```

```python
# Subplot Option 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize=(10,5))

# Plot to each different axis
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter.keys(), nut_butter.values());
ax[1,1].hist(np.random.randn(100));
```

### Which one to use? (PyPlot vs MatPlotLib Object-Oriented Method)
* When plotting something quickly, pyplot
* When plotting something more advanced, matplotlib(object-oriented)

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#Pyplot Method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
# Object-Oriented Method + Pyplot Method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);

ax.set_xlim([45,100])
```

```python
# Object-Oriented Method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="dashed")
```
